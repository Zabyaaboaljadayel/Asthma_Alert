```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
wheatherData = pd.read_excel('Lisbon,Portugal 2019-01-01 to 2020-03-31.xlsx')
wheatherData
```

```python
pollenBuf = pd.read_excel('Pollen_2019.xlsx').iloc[:,:2]
pollenBufDate = []
for i in range(len(pollenBuf)):
    pollenBufDate.append(pollenBuf.date[i].strftime("%Y-%m-%d"))
pollenBuf.date = pollenBufDate
pollenBuf.date[0]
# pollenBufDate[0]
```

```python
pollenData = pd.DataFrame()
pollenData['date'] = wheatherData['datetime']
# wheatherData['datetime'] = pollenData['date']
pollenData['pollen_intensity'] = 'good'
pollenData
```

```python
intensityBuf = pollenBuf['pollen_intensity']
datepollenBuf = list(pollenBuf['date'])
pollen_intensity = pollenData['pollen_intensity']
datepollenData = pollenData['date']

for i in range(len(pollen_intensity)):
    if datepollenData[i] in datepollenBuf:
#         print(datepollenBuf[datepollenBuf.index(datepollenData[i])],
#               datepollenData[i],
#               intensityBuf[datepollenBuf.index(datepollenData[i])])
        pollen_intensity[i] = intensityBuf[datepollenBuf.index(datepollenData[i])]
pollenData['pollen_intensity'] = pollen_intensity
```
