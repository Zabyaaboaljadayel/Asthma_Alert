```python
import pandas as pd

df = pd.read_csv('pollen data.csv')
df.head(3)
```

```python
months_en2pt = {"January":"janeiro","February":"fevereiro",
               "March":"março","April":"abril","May":"maio",
               "June":"junho","July":"julho","August":"agosto",
               "September":"setembro","October":"outubro",
               "November":"novembro", "December":"dezembro"
              }
months_pt2en = {v: k for k, v in months_en2pt.items()}
```

```python
def get_my_two_dates(cleaned_date_list):
    if len(cleaned_date_list) == 4:
        pt_month = cleaned_date_list[2]
        en_month = months_pt2en[pt_month.lower()]

        date1 = cleaned_date_list[0] +'-'+ en_month +'-'+ cleaned_date_list[-1]
        date2 = cleaned_date_list[1] +'-'+ en_month +'-'+ cleaned_date_list[-1]
        return date1,date2
    else:
        
        pt_month_1 = cleaned_date_list[1]
        pt_month_2 = cleaned_date_list[3]
        
        en_month_1 = months_pt2en[pt_month_1.lower()]
        en_month_2 = months_pt2en[pt_month_2.lower()]

        date1 = cleaned_date_list[0] +'-'+ en_month_1 +'-'+ cleaned_date_list[-1]
        date2 = cleaned_date_list[1] +'-'+ en_month_2 +'-'+ cleaned_date_list[-1]
        return date1,date2
```

```python
clean_date_string = lambda date_string: [e for e in date_string.split(' ') if e not in ['a','de']]
cleaned_date_list = clean_date_string('22 a 29 de Abril de 2019')
date1,date2 = get_my_two_dates(cleaned_date_list)
data = {'date':pd.date_range(start=date1,end=date2),
        'pollen_intensity':'something',
        'source_type':None # Complete
       }
pd.DataFrame(data)
```
